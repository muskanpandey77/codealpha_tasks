```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("samplesuperstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
region_profit = df.groupby('Region')['Profit'].sum()

region_profit.plot(kind='pie', autopct='%1.1f%%')

plt.title("Profit by Region")

plt.show()
```

```python
sns.scatterplot(x=df['Discount'], y=df['Profit'])

plt.title("Discount vs Profit")

plt.show()
```

```python
sns.countplot(x='Category', data=df)

plt.title("Number of Products by Category")

plt.show()
```

```python
plt.hist(df['Sales'])

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.barplot(x='Category', y='Profit', data=df)

plt.title("Profit by Category")

plt.show()
```

# Conclusion

1. Technology category generated high sales.

2. Some regions contributed more profit than others.

3. Higher discounts often reduced profits.

4. Data visualization made business trends easier to understand.

5. Charts and graphs improved data interpretation and storytelling.
