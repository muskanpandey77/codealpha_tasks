```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
df = df = pd.read_csv("IMDB Dataset.csv")
```

```python
df.head()
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        return "Positive 😊"
    elif analysis.sentiment.polarity < 0:
        return "Negative 😡"
    else:
        return "Neutral 😐"
```

```python
df['Predicted_Sentiment'] = df['review'].apply(get_sentiment)
```

```python
df[['review', 'Predicted_Sentiment']].head()
```

```python
df['Predicted_Sentiment'].value_counts()
```

```python
df['Predicted_Sentiment'].value_counts().plot(kind='bar')

plt.title("Sentiment Analysis Result")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")

plt.show()
```

# Conclusion

1. Most movie reviews were classified as either positive or negative.

2. TextBlob successfully analyzed sentiments from real IMDb reviews.

3. Sentiment analysis helps understand public opinion and audience reactions.

4. Visualization made it easier to identify sentiment distribution patterns.
